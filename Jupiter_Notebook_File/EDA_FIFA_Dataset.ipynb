```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("fifa_data.csv")
df
```

```python
#Basic information about the dataset
print(df.info())
```

```python
#Summary statistics of numerical columns
print(df.describe())
```

```python
#Number of missing values in each column
print(df.isnull().sum())
```

```python
#First few rows of the dataset
print(df.head())
```

```python
duplicate_rows = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_rows)
```

```python
# Histogram of Overall ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Overall'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Age vs Overall rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Overall', color='green')
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

```python
# Boxplot of Overall ratings by Preferred Foot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Preferred Foot', y='Overall', palette='Set2')
plt.title('Overall Rating by Preferred Foot')
plt.xlabel('Preferred Foot')
plt.ylabel('Overall Rating')
plt.show()
```

```python
most_players_country = df['Nationality'].value_counts().idxmax()
print("Country with the most number of players:", most_players_country)
```

```python
top_countries = df['Nationality'].value_counts().head(5)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
for i, (country, count) in enumerate(top_countries.items()):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')
plt.show()
```

```python
df['Wage'] = df['Wage'].str.replace('€', '').str.replace('K', '').astype(int)
highest_salary_player = df.loc[df['Wage'].idxmax()]
print("\nPlayer with the highest salary:")
print(highest_salary_player[['Name', 'Wage']])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Wage'], bins=20, kde=True, color='orange')
plt.title('Salary Range of Players')
plt.xlabel('Wage (in Euros)')
plt.ylabel('Frequency')
plt.show()
```

```python
df = df.dropna(subset=['Height'])
df['Height'] = (df['Height'].str.extract(r"(\d+)\s*'(\d+)").astype(int).iloc[:, 0] * 30.48 + df['Height'].str.extract(r"(\d+)\s*'(\d+)").astype(int).iloc[:, 1] * 2.54).astype(int)
tallest_player = df.loc[df['Height'].idxmax()]
print("\nThe tallest player in FIFA is:", tallest_player['Name'])
```

```python
most_players_club = df['Club'].value_counts().idxmax()
print("\nClub with the most number of players:", most_players_club)
```

```python
preferred_foot_counts = df['Preferred Foot'].value_counts()
plt.figure(figsize=(8, 6))
preferred_foot_counts.plot(kind='bar', color=['blue', 'green'])
plt.title('Preferred Foot of Players')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```
